```python
#import the necessary libraries
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#import the sample dataset from github
df = pd.read_csv('https://raw.githubusercontent.com/sh-dev-gh/Mathematics-for-Applied-AI/main/Rating_dataset.csv')
```

```python
#show the first 5 rowns
df.head()
```

```python
#representing the dataset as a matrix
matrix = df.pivot_table(index='User_id', columns='Movie_id', values = 'Ratings')
matrix.head()
```

```python
#replacing the null values in the table with zeros to be able to calculate the cosine similarity
final_matrix = matrix.fillna(0)
```

```python
#using the cosine similarity function from sklearn to calculate the similarities between the users
user_similarity = cosine_similarity(final_matrix)
user_similarity_matrix = pd.DataFrame(user_similarity, index=matrix.index, columns=matrix.index)
```

```python
#create a function that takes the user id, matrix and similarities and return which movie will be recommended
def recommend_movie(User_id,final_matrix,user_similarity_matrix):
  movie_rating = []
  #add the movie that wasn't show by user with User_id to rec_movie
  for i in range(len(final_matrix.columns)):
    if final_matrix.iloc[User_id,i]==0:
      rec_movie = final_matrix.columns[i]
      rating = 0
      m_similarities = 0

      #check the reatings of other users
      for index in range(len(user_similarity_matrix.columns)):
        if User_id != index and final_matrix.iloc[index,i] != 0:
          #calculate the rating for the movie depending on the similarities between the user with User_id and the others
          rating += final_matrix.iloc[index,i]*user_similarity_matrix.iloc[User_id,index]
          m_similarities += user_similarity_matrix.iloc[User_id,index]
      if m_similarities !=0:
        f_rating = rating/m_similarities
      movie_rating.append((rec_movie,f_rating))
  #find the movie with highest rating to recommend it
  rec_movie = max(movie_rating, key=lambda x: x[1])
  return print('Recommended movie for User ID',User_id,'will be movie ID',rec_movie[0], 'with a ranting of',rec_movie[1])
```

```python
#find which movie to recommend to user with User_id=0
recommend_movie(0,final_matrix,user_similarity_matrix)
```
